```python
from matplotlib import pyplot as mpl

subjects=['history','english','maths','science','computers','hindi']
avgscore=[73,58,63,82,91,79]
clrs=['red','green','blue','magenta','cyan','purple']

mpl.bar(subjects,avgscore,color=clrs)
mpl.show()
```

```python
from matplotlib import pyplot as mpl
import pandas

df=pandas.read_csv('covid-usa.csv')
df1=df[['month','cases']]
#print(df1)

mpl.plot(df1['month'],df1['cases'],color='blue')
mpl.scatter(df1['month'],df1['cases'],color='black')

df=pandas.read_csv('covid-usa.csv')
df2=df[['month','deaths']]
#print(df2)

mpl.plot(df2['month'],df2['deaths'],color='red')
mpl.scatter(df2['month'],df2['deaths'],color='green')

mpl.show()
```

```python
import numpy
from matplotlib import pyplot as mpl

arr=numpy.random.randint(1,50,35)
arrsq=arr**2
print(arr)
print(arrsq)

mpl.scatter(arr,arrsq)
mpl.xlabel('Numbers')
mpl.ylabel('Squares')
mpl.show()
```
